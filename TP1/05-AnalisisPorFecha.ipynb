```python
import pandas as pd
import matplotlib as mpl
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pylab

plt.style.use('default')
plt.rcParams['figure.figsize'] = [3.0,3.0]

df = pd.read_csv('data/train.csv',dtype={'tipodepropiedad':'category','ciudad':'category','provincia':'category'})
df['fecha'] = pd.to_datetime(df['fecha'])
df['anio de publicacion'] = df['fecha'].dt.year
df["mes de publicacion"] = df['fecha'].dt.month
df["dia de publicacion"] = df['fecha'].dt.day
df['preciousd'] = df['precio']/20
df["cantidad"] = 1
```

```python
columnas = ["anio de publicacion","mes de publicacion","antiguedad"]
publicado = df[columnas]
publicado = publicado.pivot_table(index='anio de publicacion', columns='mes de publicacion', values='antiguedad', aggfunc='mean')
plt.figure(figsize=(15,5))
g = sns.heatmap(publicado,  cmap="YlGnBu", linewidths=.5)
g.set_ylim(5,-0,1)
g.set_title("Antiguedad de propiedad segun Año y Mes de publicacion", fontsize=22)
g.set_xlabel("Mes de Publicaion",fontsize=18)
g.set_ylabel("Año de Publicacion", fontsize=18)
```

```python
#Heatmap de mes y tipo de prop
b = df[["anio de publicacion","mes de publicacion","tipodepropiedad","id"]]
b = b[(b["tipodepropiedad"]!="Hospedaje") & (b["tipodepropiedad"]!="Lote") & (b["tipodepropiedad"]!="Garage")]
a = b.pivot_table(index="anio de publicacion",columns="tipodepropiedad",values= "id", aggfunc= "count" ).fillna(value=0)
b = b.pivot_table(index="mes de publicacion",columns="tipodepropiedad",values= "id", aggfunc= "count" ).fillna(value=0)
b = b.transform(lambda x: (x * 100) / x.sum()).T
a = a.transform(lambda x: (x * 100) / x.sum()).T
```

## Evolucion cantidad de publicaciones por provincia a lo largo de losaños

```python
grouped = df.groupby(["provincia","anio de publicacion"]).agg({"cantidad":"sum"})
```

```python
years =df["anio de publicacion"].unique()
years.sort()
```

```python
g_years = df.groupby(["anio de publicacion"]).agg({"cantidad":"sum"})
```

```python
g_years
```

```python
x =years
y= [g_years.iloc[0][0],g_years.iloc[1][0],g_years.iloc[2][0],g_years.iloc[3][0],g_years.iloc[4][0]]
plt.rcParams['figure.figsize'] = [18.0,18.0]
plt.title("Evolucion de publicaciones", fontsize=22)
plt.xlabel("Año de publicacion", fontsize=15)
plt.ylabel("Cantidad de publicaciones", fontsize=15)

plt.stackplot(x,y, labels=["Cantidad de Anuncios"] )
plt.legend(loc = "upper left")
plt.show()
```

Se puede apreciar que del año 2015 al año 2016, la cantidad de anuncios casi
se duplicaron en comparacion a la año anterior. 
Del año 2012 al 2015 la evolucion era casi lineal.
