Se realizará un análisis sobre la proporcion en los distintos tipos de propiedad que se venden en zonaprop.com.

```python
import pandas as pd

df = pd.read_csv("../../../train.csv", sep=",")

df['tipodepropiedad'].value_counts()
```

Al parecer, zonaprop es un servicio dominado por la compra y venta de propiedades del tipo "vivienda", donde la gente reside. Llamaremos viviendas a todas las propiedades que sean del tipo Casa, Apartamento o Casa en condominio.

Vease el siguiente gràfico comparando la cantidad de publicaciones de vivienda contra no viviendas.

```python
df['esVivienda'] = df['tipodepropiedad'].map(lambda x: x in ['Casa', 'Apartamento', 'Casa en condominio'])
df.head(10)
```

```python
import seaborn as sns
sns.barplot(x= 'esVivienda',data=df)
```
